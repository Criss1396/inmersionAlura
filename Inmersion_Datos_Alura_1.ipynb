```python
top_10_barrios = ["Usaquén","Zona Noroccidental","Bosa","Kennedy","Cedritos",
                  "Barrios Unidos","Engativa","Suba","Santa Barbara","Chapinero"]
```

```python
promedios_barrios = []
for barrio in top_10_barrios:
  promedio = inmuebles[inmuebles.Barrio == barrio].Area.mean()
  promedios_barrios.append(promedio)

promedios_barrios
```

```python
promedios_barrios
```

```python
# Obtenemos los promedios de área por barrio
promedios_barrios = []
for barrio in top_10_barrios:
  promedio = inmuebles[inmuebles.Barrio == barrio].Area.mean()
  promedios_barrios.append([barrio, promedio])

promedios_barrios = pd.DataFrame(promedios_barrios, columns=["Barrio", "Promedio de área"])

#grafico de barras
promedios_barrios.plot.bar(x="Barrio")


```

```python
# Obtenemos las medianas de área por barrio
mediana_barrios = []
for barrio in top_10_barrios:
  mediana = inmuebles[inmuebles.Barrio == barrio].Area.median()
  mediana_barrios.append([barrio, mediana])

mediana_barrios = pd.DataFrame(mediana_barrios, columns=["Barrio", "Mediana de área"])

#grafico de barras
mediana_barrios.plot.bar(x="Barrio")
```

```python
mediana_barrios
```

```python
from pandas.core.reshape.merge import merge
# Obtenemos los valores minimo y maximo de área por barrio
#primero hallamos los minimos
min_barrios = []
for barrio in top_10_barrios:
  min = inmuebles[inmuebles.Barrio == barrio].Area.min()
  min_barrios.append([barrio, min])

min_barrios = pd.DataFrame(min_barrios, columns=["Barrio", "Valor minimo de área"])

#Valor maximo
max_barrios = []
for barrio in top_10_barrios:
  max = inmuebles[inmuebles.Barrio == barrio].Area.max()
  max_barrios.append([barrio, max])

max_barrios = pd.DataFrame(max_barrios, columns=["Barrio", "Valor maximo de área"])

#unimos
min_max_barrios = pd.merge(min_barrios, max_barrios, on="Barrio")

#grafico de barras
min_max_barrios.plot.bar(x="Barrio", subplots=True, )
```
